```python
from googletrans import Translator
translator = Translator()

#Codes for Different Languages are Listed below


#Tamil - ta
#Malayalam - ml
#Hindi - hi
#Telegu - te
#Kannada - kn
#Marathi - mr
#French - fr
#Spanish - es
#German - de
#Bengali - bn



text = input("Enter the text you want to translate: ")
lang = input("Enter the Language you want to translate: ")
code = input("Enter the code of the language: ")

translated_text = translator.translate(text, dest=code)
print("\nThe translation of the text [", text,"]","to",lang,"is",translated_text.text,".")
```
